```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

## Temel Çizim İşlemler

```python
plt.plot([1,2,3,4],[1,4,9,16]) # x ekseni ve y ekseni değerlerini girdik
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]

plt.plot(x,y)
plt.show()
#böyle de tanımlayabiliriz
```

```python
plt.title("ilk grafik")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
# normalde çentikler oto olarak yerleştirilir(1.0,1.5,2.0,2.5...4.0)
# kendimiz de çentiklerin yerini tanımlayabiliriz

plt.xticks([1,2,3,4]) #küsuratları da gösterebilir
plt.yticks([1,4,9,16])
plt.title("ilk grafik")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

## Lejant Koyma

```python
plt.plot(x,y,label="x^2", color="green")
plt.xticks([1,2,3,4,5]) 
plt.yticks([1,4,9,16,25])
plt.title("ilk grafik")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend() # bu denmek zorunda
plt.show()
# sağ üstteki şeyin adı lejant
```

```python
#linewidth ekleme
plt.plot(x,y,label="x^2",color="blue",linewidth=2,linestyle="--",marker=".")
plt.xticks([1,2,3,4,5]) 
plt.yticks([1,4,9,16,25])
plt.title("ilk grafik")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

## Aynı grafikte birden fazla çizim yapmak

```python
plt.plot(x,y,label="x^2",color="green",linewidth=2,linestyle="--",marker=".")
plt.xticks([1,2,3,4,5]) 
plt.yticks([1,4,9,16,25])
plt.title("ilk grafik")
plt.xlabel("x axis")
plt.ylabel("y axis")

# 2.çizgimizi çiziyoruz
x2=np.arange(0,5,0.5) # numpyın arrange fonksiyonu 0 ile 5 arasında 0.5luk aarry oluştur demek
plt.plot(x2,x2*2,label="2*x" ,color="red",linewidth=2,marker=".")

plt.legend()
plt.show()
```

## !! Grafiği bilgisayara kaydetme

```python
plt.plot(x,y,label="x^2",color="green",linewidth=2,linestyle="--",marker=".")
plt.xticks([1,2,3,4,5]) 
plt.yticks([1,4,9,16,25])
plt.title("ilk grafik")
plt.xlabel("x axis")
plt.ylabel("y axis")

# 2.çizgimizi çiziyoruz
x2=np.arange(0,5,0.5) # numpyın arrange fonksiyonu 0 ile 5 arasında 0.5luk aarry oluştur demek
plt.plot(x2,x2*2,label="2*x" ,color="red",linewidth=2,marker=".")

plt.legend()

plt.savefig("ilkGrafiğim.png",dpi=300)

plt.show()
```

## Barchart

```python
x=["ankara","istanbul","izmir"]
y=[120,178,87]
plt.bar(x,y)
plt.show()
```

```python
#eğer çubukların bir tanesini diğerlerinden farklılaştırmak için içini boyamak istersek;
x=["ankara","istanbul","izmir"]
y=[120,178,87]

cubuklar=plt.bar(x,y)
cubuklar[1].set_hatch('/')
cubuklar[2].set_hatch('.')
plt.show()
```

## Detaylı Örnekler

```python
dataset={
    "Year":[1990,1991,1992,1993,1994],
    "Canada":[1.77,1.92,1.89,1.59,1.66],
    "France":[4.59,4.67,4.87,5.11,5.22],
    "Germany":[2.22,2.31,2.30,2.25,2.20],
    "Türkiye":[5.23,4.66,4.32,4.40,4.23],    
}
dfGas=pd.DataFrame(dataset)
```

```python
plt.title("Petrol Fiyatları")
plt.plot(dfGas['Year'],dfGas['Canada'],"b-",label="Canada")
# x ekseni yıl,y ekseni kanada olacak
# b- anlamı: mavi çizgi oluştur

plt.xlabel("yıl")
plt.ylabel("dolar")

plt.legend()
plt.show()
```

```python
plt.title("Petrol Fiyatları")
plt.plot(dfGas['Year'],dfGas['Canada'],"b-",label="Kanada")
# x ekseni yıl,y ekseni kanada olacak
# b- anlamı: mavi çizgi oluştur
plt.plot(dfGas['Year'],dfGas['Türkiye'],"r-",label="Türkiye")
plt.plot(dfGas['Year'],dfGas['Germany'],"g.-",label="Almanya")
plt.plot(dfGas['Year'],dfGas['France'],"y.-",label="Fransa")


plt.xlabel("yıl")
plt.ylabel("dolar")

plt.legend()
plt.show()
```

## Figure Size

* Tablonun büyüklüğünü değiştirebiliriz
* plt.figure(figsize=(XWidth,YWidth))

```python
plt.figure(figsize=(11,5))
plt.title("Petrol Fiyatları")
plt.plot(dfGas['Year'],dfGas['Canada'],"b-",label="Kanada")
# x ekseni yıl,y ekseni kanada olacak
# b- anlamı: mavi çizgi oluştur
plt.plot(dfGas['Year'],dfGas['Türkiye'],"r-",label="Türkiye")
plt.plot(dfGas['Year'],dfGas['Germany'],"g.-",label="Almanya")
plt.plot(dfGas['Year'],dfGas['France'],"y.-",label="Fransa")


plt.xlabel("yıl")
plt.ylabel("dolar")

plt.legend()
plt.savefig("ikinciGrafigim.png", dpi=300)
plt.show()
```
