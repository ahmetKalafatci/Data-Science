### Recommendation Systems

Örnek uygulama: Movie Recommendation Software

```python
import pandas as pd
import numpy as np 
```

```python
column_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv('users.data', sep='\t',names=column_names) # sep='\t' tab ile ayrıldığını gösterir
#csv dosyamızda sütun adı yazmadığı için biz veriyoruz
```

```python
df.head()
```

```python
len(df) #kaç kayıt olduğunu öğrenelim
```

```python
# şimdi diğer dosyamızı yükleyelim
movie_titles=pd.read_csv("movie_id_titles.csv")
movie_titles.head()
```

```python
len(movie_titles)
```

```python
#şimdi pandas kullanarak merge(birleştirme) işlemi yapıcaz
df=pd.merge(df,movie_titles, on="item_id")
# item_id'ye göre join yap dedik
df.head()
```

```python
# artık tek dataframe üstünden gidicez 2 tabloyu item_id üzerinden birleştirdik
```

### Recommendation Sistemimizi Kuruyoruz:

```python
# öncelikle exceldeki pivot tablo benzeri bir yapı kuruyoruz
# bu yapıya göre her satır bir kullanıcı olacak şekilde(yani df'nin index'i user_id olacak)
# sütunların film isimleri (yani title sütunu) olacak
# tablo içinde de rating değerleri olacak şekilde bir df oluşturuyoruz

moviemat=df.pivot_table(index='user_id', columns='title',values='rating')
moviemat.head()
```

```python
# dataframe'imizi oluşturduk

# şimdi Star Wars(1977) filminin user ratinglerine bakalım:
```

```python
starwars_user_ratings=moviemat["Star Wars (1977)"]
starwars_user_ratings.head()
```

```python
# amacımız kullanıcıya star wars filmine benzer bir öneri yapmak
```

```python
# corwith() metodunu kullanarak star wars filmi ile korelasyonları hesaplatalım:
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars
```

#### Bazı kayıtlarda boşluklar olduğu için hata veriyor similor_to_starwars bir seri, biz bunu corr_starwars isimli bir dataframe'e dönüştürelim ve NaN kayıtları temizleyip bakalım


```python
corr_starwars=pd.DataFrame(similar_to_starwars, columns=['Correlation']) #dfye dönüştürdük
corr_starwars.dropna(inplace=True) #bozuk verileri silip güncelledik
corr_starwars.head()
```

* mesela verilere bakarak star wars filminin örneğin 1-900 (1994) filmiyle uyumunun çok kötü olduğunu görebiliriz.


#### Elde ettiğimiz dataframe'i sıralayalım ve görelim bakalım star wars'a en yakın tavsiye edeceği film neymiş:

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

#### Görüldüğü gibi alakasız sonuçlar çıktı. bunun nedeni filmlerin çok az oy alması. bu durumu düzeltmek için 100'den az oy alan filmeri eleyelim ve bu amaçla ratings isimli bir df oluşturalım burada her filmin kaç tane oy adlığını yani oy sayısını tutalım 

```python
df.head()
```

```python
# timestamp sütununa ihtiyacımız yok silelim
df.drop(['timestamp'],axis=1)
```

```python
#her filmin ortalama(mean value) rating değerini bulalım
ratings=pd.DataFrame(df.groupby('title')['rating'].mean()) 
#title'a göre gruplayıp rating sütununun ortalamasını alıyoruz ve bunu ratings isimli df'ye dönüştürür

#büyükten küçüğe sıralayıp bakalım
ratings.sort_values('rating',ascending=False).head()
```

#### Dikkat!! Bu ortalaamalar hesaplanırken kaç oy aldığına bakmadık o yüzden bilmediğimiz filmler çıktı

```python
#şimdi her filmin oy sayısını bulalım
ratings['rating_oy_sayisi']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
#şimdi en çok oy alan filmeri sıralayalım
ratings.sort_values('rating_oy_sayisi',ascending=False).head()
```

```python
# tekrar esas amacımıza dönelim ve corr_starwars df'imize rating_oy_sayisi sütununu ekleylim(join ile)
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['rating_oy_sayisi'])
corr_starwars.head() 
#join ile rating_oy_sayisi sütununu corr_starwars df sine ekledik
```

### Ve sonuç:

```python
corr_starwars[corr_starwars['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False).head(10)
#oy sayısı 100den fazla olanları baz alıp benzerliğe göre sıraladık
```
