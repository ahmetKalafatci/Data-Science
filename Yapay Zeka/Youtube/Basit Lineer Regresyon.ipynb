```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Veri setinin eklenmesi

```python
#dataset=pd.read_csv("dosya_adı.csv") dosya olunca böyle ama dosya bizde olmadığı için dataseti elimle yazıcam
dataset={
    "YearsExperince":[1.1,1.3,1.5,2.0,2.2,2.6,3.0,3.2,3.2,3.7,3.9,4.0,4.0,4.1,4.5,5.1],
    "Salary":[39343,48205,37332,43523,39991,58888,60150,54445,64445,57819,63281,60123,65789,74363,62413,83881]
}
df=pd.DataFrame(dataset)
#verisetinden x ve y yi yani bağımlı ve bağımsız değişkeni çıkarmak için iloc kullanıcaz
X=df.iloc[:,:-1].values #özellikler/ bağımsız değişken
y=df.iloc[:,-1].values # bağımlı değişken
```

## Veri setinin öğrenme ve test verisi olmak üzere ikiye bölünmesi

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=0)
```

## Basit linner regresyon modelinin öğrenme verisinde öğrenmesini gerçekleştirmesi

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
#öğrenme için her zaman fit fonksiyonunu kullanırız 
```

## Test verisi üzerinde tahminlerin yapılması

```python
y_pred=regressor.predict(x_test)
```

## Öğrenme sonuçlarının görselleştirilmesi

```python
plt.scatter(x_train,y_train,color="red") #öğrenme verisinin noktaları yani maaş-deneyim ikilileri
plt.plot(x_train,regressor.predict(x_train),color="blue") #regresyon modelimizin bulduğu doğrı
plt.title("Maaş ve Deneyim ( Öğrenme Verisi ) ") #başlık
plt.xlabel("Deneyim") #x doğrusu deneyim
plt.ylabel("Maaş") # y doğrusu maaş
plt.show() #göster
```

## Test sonuçlarının görselleştirilmesi

```python
plt.scatter(x_test,y_test,color="red") 
plt.plot(x_train,regressor.predict(x_train),color="blue") # bu doğru sabittir çünkü öğrenme verisinde öğrenme yapılmıştır ve testte değiştirilmez
plt.title("Maaş ve Deneyim ( Test Verisi ) ") #başlık
plt.xlabel("Deneyim") #x doğrusu deneyim
plt.ylabel("Maaş") # y doğrusu maaş
plt.show() #göster
```

## Tahmin Yapmak

```python
print(regressor.predict([[12]])) #12 yıl çalışna birinin maaşını tahmin ettirdik
#tahmin fonksiyonu her zaman bizdedn 2 boyutlu dizi ister o yüzden çift parantez kullanılıd
#[[12]]-> 2d
#[12]-> 1d
```

## Formülün bulunması

```python
print(regressor.coef_) #b1 katsayısı
print(regressor.intercept_) #b0 katsayısı
```
